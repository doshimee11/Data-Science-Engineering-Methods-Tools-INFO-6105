```python
depend_var = bos_hous['MV']
categ_var = bos_hous[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PT','B','LSTAT']]
lm = linear_model.LinearRegression()
lm.fit(categ_var,depend_var)

categ_var = sm.add_constant(categ_var)
lm = sm.OLS(depend_var,categ_var).fit()
print(lm.summary())
```

**Apply logistic regression on the Loan dataset with "Decision" as the dependent variable. Do the necessary one-hot encoding for the categorical variables and discard the irrelevant variables. Use whatever libraries are appropriate when coding in Python.**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import statsmodels.api as sm
from statsmodels.api import OLS
import seaborn as sns

loan = pd.read_excel('loan.xlsx', index_col = None)
loan.nunique()
loan2 = pd.get_dummies(loan, columns=['Sex','Res_status','Telephone','Occupation','Liab_ref','Acc_ref','Job_status'])
print(loan2.head())
loan2['Decision'] = loan2['Decision'].replace(['accept','reject'],[1,0])

lr_model = LogisticRegression(solver='liblinear', random_state=0)
dep_var = loan2['Decision']
cat_var = loan2.drop(['Decision'],axis=1)
lr_model.fit(cat_var, dep_var)

cat_var = sm.add_constant(cat_var)
lr_model = sm.OLS(dep_var, cat_var).fit()
lr_model.summary()
```

```python
lr_model1 = LogisticRegression(solver='liblinear', random_state=0)
dep_var1 = loan2['Decision']
cat_var1 = loan2.drop(['Decision','Sex_M','Time_at_address','Res_status_rent','Telephone_not_given',
                      'Occupation_creative_','Occupation_executive','Occupation_manager',
                      'Occupation_office_st','Occupation_productio','Occupation_sales','Occupation_unemploye',
                      'Liab_ref_f','Acc_ref_given','Acc_ref_oth_inst_','Job_status_military',
                       'Job_status_retired','Job_status_student','Job_status_unemploye'],axis=1)
lr_model1.fit(cat_var1, dep_var1)

cat_var1 = sm.add_constant(cat_var1)
lr_model1 = sm.OLS(dep_var1, cat_var1).fit()
lr_model1.summary()
```
